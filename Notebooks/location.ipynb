````python
import http.client
import json
import urllib.parse
````

````python
conn = http.client.HTTPSConnection("google-map-places.p.rapidapi.com")
headers = {
    'x-rapidapi-key': "7ee1c65373msh181b4a01f21d239p1e5ff7jsnc2f07e071e6d",
    'x-rapidapi-host': "google-map-places.p.rapidapi.com"
}
company = 'apple'
location = 'new york'
query_params = {
    'query': f'{company} headquarters in {location}',
    'radius': 50000,
    'language': 'en',
    'type': 'establishment|point_of_interest',

}
url_safe_query = urllib.parse.urlencode(query_params)
url_safe_query
````

````python
conn.request("GET", f"/maps/api/place/textsearch/json?{url_safe_query}", headers=headers)
````

````python
output_dict = json.loads(data)
output_dict
````

````python
results = output_dict.get('results')
results
````

````python
for r in results:
    types = r.get('types',[])
    if any(['electronics_store' in types, 'store' in types]):
        print('not got it nigga !')
    else:
        print('got it nigga !')
        geometry = r.get('geometry', {})
        lat_long = geometry.get('location', {})
        job_info = {
            # place_id
            'place_id': r.get('place_id', ''),
            'name': r.get('name', ''),
            'Address': r.get('formatted_address', ''),
            'opening_hours': r.get('opening_hours', {}).get('open_now', False),
            # get latitude and longitude
            'location': lat_long,
            # get northeast and southwest
            'northeast': geometry.get('viewport', {}).get('northeast', {}),
            'southwest': geometry.get('viewport', {}).get('southwest', {}),
            # get types
            'type': types
        }
        print(job_info)
        break
````

## Using 'Google Place Autocomplete and Place Info'
Hard-limit : 1000 requests/MO

````python
conn = http.client.HTTPSConnection("google-place-autocomplete-and-place-info.p.rapidapi.com")
headers = {
    'x-rapidapi-key': "7ee1c65373msh181b4a01f21d239p1e5ff7jsnc2f07e071e6d",
    'x-rapidapi-host': "google-place-autocomplete-and-place-info.p.rapidapi.com"
}
query_params = {
    'input': f'apple headquarters in india'
}
# get the url-safe-query
url_safe_query = urllib.parse.urlencode(query_params)
````

````python
# request connection to the auto-complete endpoint
conn.request("GET", f"/maps/api/place/autocomplete/json?{url_safe_query}", headers=headers)
````

````python
res = conn.getresponse()
data = res.read()
````

````python
output_dict = json.loads(data)
````

````python
output_dict
````

````python
import json
predictions = output_dict.get('predictions')
for p in predictions:
    types = p.get('types', [])
    if('establishment' in types and 'point_of_interest' in types):
        description = p.get('description','')
        place_id = p.get('place_id', '')
        name = p.get('structured_formatting', {}).get('main_text', '')
        address = p.get('structured_formatting', {}).get('secondary_text', '')
        
        job_info = {
            'place_id': place_id,
            'name': name,
            'address': address,
            'description': description,
            'types': types,
        }
        print(job_info,'\n')
````

````python
for r in results:
    container = []
    # get id
    place_id = r.get('id', ''),
    address = r.get('title', ''),
    latitude = r.get('latitude', '')
    longitude = r.get('longitude', '')
    country_code = r.get('country_code')
    is_place = r.get('is_place')
    location_info = {
        'place_id': place_id,
        'address': address,
        'latitude': latitude,
        'longitude': longitude,
        'country_code': country_code,
        'is_place': is_place
    }
    container.append(location_info)
    print(location_info,'\n')
````

````python
def extract_valuable_information(output_dict):
    extracted_data = []
    data = output_dict.get('data', {})
    # access the results list from the data
    results = data.get("results", [])
    
    for record in results:
        location_info = {
            "id": record.get("id", "N/A"),
            "name": record.get("name", "N/A"),
            "about": record.get("about", "N/A"),
            "phone_number": record.get("phone_number", "N/A"),
            "address": record.get("address", "N/A"),
            "zone": record.get("zone", "N/A"),
            "country_code": record.get("country_code", "N/A"),
            "url": record.get("url", "N/A"),
            # "reviews": record.get("reviews", "N/A"),
            "rating": record.get("rating", "N/A"),
            "tag": record.get("tag", "N/A"),
            # "cover_photo": record.get("cover_photo", "N/A"),
            # "photo_count": record.get("photo_count", "N/A"),
            "status": record.get("status", "N/A"),
            "opening_status": record.get("opening_status", "N/A"),
            "website": record.get("website", "N/A"),
            "plus_code": record.get("plus_code", "N/A"),
            "latitude": record.get("latitude", "N/A"),
            "longitude": record.get("longitude", "N/A"),
        }
        extracted_data.append(location_info)

    return extracted_data
````
