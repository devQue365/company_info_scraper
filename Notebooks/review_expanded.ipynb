### For selecting the first proxy which returns successful status

```python
async def check_proxy(proxy):
    # test url
    test_url = 'https://httpbin.org/ip'
    # get header with random user agent
    user_agent = ua.random
    headers = {
        'User-Agent': user_agent
    }
    try:
        # create an asynchronous session
        async with aiohttp.ClientSession() as session:
            async with session.get(test_url, proxy = proxy, timeout = 10, headers = headers) as res:
                if (res.status == 200):
                    print("Proxy worked !!!!")
                    return proxy
                else:
                    print("Proxy did not work !!!!")
                    return None
    except:
        print("Could not establish session !")
        return None
    
    
```

### Function to get the page we have searched for

```python
async def request_page(proxy, url):
    user_agent = ua.random
    headers = {
        'User-Agent': user_agent
    }
    try:
        async with aiohttp.ClientSession() as session:
            async with session.get(url, proxy = proxy, headers = headers, timeout = 10) as res:
                html = await res.text()
                return html
    except:
        print("Could not reach the specified url !!!")
        return None
            
```

### This is the main logic in which we will be trying all proxies in parallel

```python
async def request_with_rotation(proxypool, url):
    # run all the proxies in parallel i.e. concurrent way
    jobs = [asyncio.create_task(check_proxy(proxy)) for proxy in proxypool]
    # check status of proxies
    done, pending = await asyncio.wait(jobs, return_when=asyncio.FIRST_COMPLETED)
    # for the proxy which got selected
    selected_proxy = None
    # check the done segment
    for job in done:
        result = job.result()
        if result:
            selected_proxy = result
            break # no need to check further
    # terminate the pending jobs
    for job in pending:
        job.cancel()
        with suppress(CancelledError):
            await job
            
    # check if no proxy is selected i.e. all are out of service
    if not selected_proxy:
        print("None of the proxy is in working state ...")
        return None
    # --------- Request the page -------
    html_content = await request_page(selected_proxy, url)
    return html_content
```
